```python
from astropy.io import fits
from astropy.wcs import WCS
import matplotlib.pyplot as plt
import numpy as np

import btk
import btk.plot_utils
import btk.survey
import btk.draw_blends
import btk.catalog
import btk.sampling_functions
import astropy.table
import galsim
import os
```

```python
save_path_blends = 'output/blends/Rubin/blended.npy'
save_path_isolated = 'output/blends/Rubin/isolated.npy'
#these are my save paths so just change them
```

```python
blends = np.load(save_path_blends)
isolated = np.load(save_path_isolated)
#Loads in both the files
```

```python
num_blends = 19 #this is just the number of blends you have -1. It should all work after you set this value
n = 0
blendsplus = np.zeros((3, (num_blends+1), 64, 64))
while n <= num_blends:
    blendsplus[0,n] = blends[n,1]
    blendsplus[1,n] = isolated[n,0,0]
    blendsplus[2,n] = isolated[n,1,0]
    n = n+1
#this is what puts the blends and isolated into the format you can use, just a simple while loop tbh
```

```python
plt.imshow(blendsplus[1,18])
np.shape(blendsplus)
#This is the file that has all the blends as well as the isolated ones. 0 is the blends then 1 and 2 are the isolated.
#1 is the centred galaxy if you need to know
```

```python
isolated_and_blends = "output/blends_and_isolated" #change this to the savepath you want
np.save(isolated_and_blends,blendsplus) #this just saves it to the path you set, you can change name if you want
```
